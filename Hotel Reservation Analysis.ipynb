```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df['arrival'] = df['arrival_date_day_of_month'].astype(str)+"-"+df['arrival_date_month']+"-"+df['arrival_date_year'].astype(str)
df
```

```python
df.drop(['arrival_date_day_of_month', 'arrival_date_month', 'arrival_date_year', 'arrival_date_week_number'], axis=1, inplace=True)
df
```

```python
df['arrival'] = df['arrival'].astype('datetime64[ns]')
```

```python
df['children'] = df['children'] + df['babies'].astype('float64')
```

```python
df['lodgings'] = df['stays_in_week_nights'] + df['stays_in_weekend_nights']
```

```python
df.drop(['babies', 'company'], axis=1, inplace=True)
df
```

```python
df['country'] = df['country'].fillna(method='bfill')
df['agent'] = df['agent'].fillna((df['agent'].median()))
df['children'] = df['children'].fillna((df['children'].mode()))
```

```python
df.isnull().sum()
```

```python
corrmat = df.corr()
fig, ax = plt.subplots(figsize=(20,15))
sns.heatmap(corrmat, annot=True, fmt='.2f', ax=ax, cmap='RdYlGn')
plt.show()
```

```python
#Using Boxplot to understand the distribution of the data more
fig, ax = plt.subplots(figsize=(20,8), ncols=5)

sns.boxplot(x='is_canceled', y='lead_time', data=df, ax=ax[0])
ax[0].set_title('Lead Time vs Target')

sns.boxplot(x='is_canceled', y='previous_cancellations', data=df, ax=ax[1])
ax[1].set_title('Previous Cancellations vs Target')

sns.boxplot(x='is_canceled', y='required_car_parking_spaces', data=df, ax=ax[2])
ax[2].set_title('Parking Spaces vs Target')

sns.boxplot(x='is_canceled', y='total_of_special_requests', data=df, ax=ax[3])
ax[3].set_title('Special Requests vs Target')

sns.boxplot(x='is_canceled', y='booking_changes', data=df, ax=ax[4])
ax[4].set_title('Booking Changes vs Target')

plt.show()
```
